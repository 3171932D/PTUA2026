```python
import math
def sphereVolume(radius):
    return (4*math.pi*(radius**3))/3
```

```python
bookRRP = 24.95
bookWholesalePrice = bookRRP * 0.4
def totalWholesaleCost(numberOfCopies):
    bookTotalCost = bookWholesalePrice * numberOfCopies
    shippingTotalCost = 3
    if numberOfCopies > 1:
        return bookTotalCost + shippingTotalCost + ((numberOfCopies-1) * 0.75)
    else:
        return bookTotalCost + shippingTotalCost
```

```python
print(totalWholesaleCost(2))
```

```python
import datetime, time
startTime = datetime.timedelta(hours=6, minutes=52)
easyPace =  datetime.timedelta(minutes=8, seconds=15)
tempoPace = datetime.timedelta(minutes=7, seconds=12)
def totalTime(easyPaceMiles, tempoPaceMiles):
    totalTime = datetime.timedelta()
    totalEasyPace = datetime.timedelta()
    totalTempoPace = datetime.timedelta()    
    for i in range(easyPaceMiles):
       totalEasyPace = totalEasyPace + easyPace
    print("easyPace total: " + str(totalEasyPace))
    for i in range(tempoPaceMiles):
        totalTempoPace = totalTempoPace + tempoPace
    print("tempoPace total: " + str(totalTempoPace))
    totalTime = totalEasyPace + totalTempoPace
    print("Total Time: " + str(totalTime))
    print("Start Time: " + str(startTime))
    print("End Time: " + str(startTime + totalTime))
print(startTime, easyPace, tempoPace)
```

```python
totalTime(2, 3)
```
